```python
def generate_symmetric_distance_matrix(size):
    # Generate a random square matrix
    matrix = np.random.rand(size, size)
    
    # Make it symmetric by taking the average with its transpose
    sym_matrix = (matrix + matrix.T) / 2
    
    # Set the diagonal to zero (distance from an element to itself)
    np.fill_diagonal(sym_matrix, 0)
    
    return sym_matrix

# Specify the size of the matrix
matrix_size = 3  # Replace with your desired size

# Generate a random symmetric square distance matrix
distance_matrix = generate_symmetric_distance_matrix(5)

print(distance_matrix)
```
